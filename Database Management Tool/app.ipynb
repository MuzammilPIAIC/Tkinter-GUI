```python
import tkinter as tk

from tkinter import *
import tkinter.ttk as ttk
from PIL import ImageTk, Image
from tkinter import messagebox as mb
import mysql.connector
    

root=tk.Tk()
root.geometry("800x600")
root.resizable(width=0, height=0)
root.title('Security Management Software')

frame=tk.Frame(root,bg='lightblue')
frame.place(relheight=1,relwidth=1)

# function to change properties of button on hover
def changeOnHover(button, colorOnHover, colorOnLeave):
  
    # adjusting backgroung of the widget
    # background on entering widget
    button.bind("<Enter>", func=lambda e: button.config(
        background=colorOnHover))
  
    # background color on leving widget
    button.bind("<Leave>", func=lambda e: button.config(
        background=colorOnLeave))

def add_new_db_button():
    user_name = mystring.get()
    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password=""
    )

    mycursor = mydb.cursor()
    db_cm = 'CREATE DATABASE '+user_name
    mycursor.execute(db_cm)
    Home()
    
        
def add_new_db():
    for item in frame.place_slaves():
        item.place_forget()
    global mystring
    mystring = StringVar()
    text1=tk.Label(frame,text='Enter Your Database Name',font=("Helvatical bold", 15),bg='lightblue')
    text1.place(relx=0.5,rely=0.35, anchor="center")
    
    entry_feild = tk.Entry(frame, textvariable = mystring, font=("Helvatical bold", 20)) #entry textbox  
    entry_feild.place(relx=0.5,rely=0.5, anchor="center", width=250,height=40)
    
    user_submit= tk.Button(frame, text="Submit", width=15, height=2, command=add_new_db_button)
    user_submit.place(relx=0.5,rely=0.7, anchor="center")

def call():
    res=mb.askquestion('Confirmation', 'Do you really want to delete?')
    if res == 'yes' :
        delete_db()
    else :
        show_existing_db()

def call2():
    res=mb.askquestion('Confirmation', 'Do you really want to delete?')
    if res == 'yes' :
        delete_table()
    else :
        show_existing_table()

def show_existing_db():
    for item in frame.place_slaves():
        item.place_forget()
    frame2=tk.Frame(frame,bg='lightblue')
    frame2.place(relheight=.6,relwidth=.6,relx=0.2,rely=0.2)
    
    label=tk.Label(frame,text='All databases',font=("Helvatical bold", 15),bg='lightblue')
    label.place(relx=0.5,rely=0.1,anchor="center")
        
    scrollbary = Scrollbar(frame2, orient=VERTICAL)
    global tree
    tree = ttk.Treeview(frame2, columns=("Names"), height=200, selectmode="extended",
                    yscrollcommand=scrollbary.set)
    scrollbary.config(command=tree.yview)
    scrollbary.pack(side=RIGHT, fill=Y)
    tree.heading('Names', text="Names", anchor=W)
    #tree.heading('Fingerprints', text="Fingerprints", anchor=W)
    tree.column('#0', stretch=NO, minwidth=0, width=0)
    #tree.column('#1', stretch=NO, minwidth=0, width=120)
    tree.pack()

    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password=""
    )

    mycursor = mydb.cursor()

    mycursor.execute("SHOW DATABASES")
    db_list = []
    for x in mycursor:
        db_list.append(x[0])
    #print(x[0])

    for i in db_list:
        tree.insert('', 'end', values=i)

    
    
    back_buttom= tk.Button(frame, text="Back", width=20, height=2, command=Home)#, command=Home
    back_buttom.place(relx=0.1,rely=0.815)
    
    delete_buttom= tk.Button(frame, text="Delete", width=20, height=2, command=call)
    delete_buttom.place(relx=0.7,rely=0.815)

    delete_buttom= tk.Button(frame, text="Modify Database", width=20, height=2, command=inside_db)
    delete_buttom.place(relx=0.5,rely=0.85, anchor="center")


def delete_db():
    
    # Get selected item to Delete
    #global values
    selected_item = tree.selection()[0]
    values = tuple(tree.item(selected_item)['values'])

    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password=""
    )

    mycursor = mydb.cursor()
    d_db_name = 'DROP DATABASE '+values[0]
    mycursor.execute(d_db_name)
    # print(values)
    show_existing_db()

def delete_table():
    
    # Get selected item to Delete
    #global values
    selected_item = tree2.selection()[0]
    values24 = tuple(tree2.item(selected_item)['values'])

    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database=values2[0]
    )

    mycursor = mydb.cursor()
    d_db_name = 'DROP TABLE '+values24[0]
    mycursor.execute(d_db_name)
    # print(values)
    show_existing_table()

def inside_db():
    for item in frame.place_slaves():
        item.place_forget()
    global values2
    selected_item = tree.selection()[0]
    values2 = tuple(tree.item(selected_item)['values'])

    label=tk.Label(frame,text='Database name '+str(values2[0]),font=("Helvatical bold", 15),bg='lightblue')
    label.place(relx=0.5,rely=0.15, anchor="center")

    login_button= tk.Button(frame, text="Create Table", width=15, height=2, command=add_new_table)# command=add_new_db
    login_button.place(relx=0.6,rely=0.5)

    add_modify_button= tk.Button(frame, text="Existed Tables", width=15, height=2, command=show_existing_table)
    add_modify_button.place(relx=0.3,rely=0.5)

    changeOnHover(login_button, "lightgreen", "white")
    changeOnHover(add_modify_button, "lightgreen", "white")


def add_new_table():
    for item in frame.place_slaves():
        item.place_forget()
    global mystring
    mystring = StringVar()
    text1=tk.Label(frame,text='Enter Your Table Name',font=("Helvatical bold", 15),bg='lightblue')
    text1.place(relx=0.5,rely=0.35, anchor="center")
    
    entry_feild = tk.Entry(frame, textvariable = mystring, font=("Helvatical bold", 20)) #entry textbox  
    entry_feild.place(relx=0.5,rely=0.5, anchor="center", width=250,height=40)
    
    user_submit= tk.Button(frame, text="Submit", width=15, height=2, command=add_new_table_button)
    user_submit.place(relx=0.5,rely=0.7, anchor="center")

def add_new_table_button():
    table_name = mystring.get()
    print("table Name is ", table_name)
    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database=values2[0]
    )

    mycursor = mydb.cursor()
    db_cm = 'CREATE TABLE '+table_name+' (id INT(255))'
    mycursor.execute(db_cm)
    inside_db()


def show_existing_table():
    for item in frame.place_slaves():
        item.place_forget()
    frame2=tk.Frame(frame,bg='lightblue')
    frame2.place(relheight=.6,relwidth=.6,relx=0.2,rely=0.2)
    
    label=tk.Label(frame,text='All Tables',font=("Helvatical bold", 15),bg='lightblue')
    label.place(relx=0.5,rely=0.1,anchor="center")
        
    scrollbary = Scrollbar(frame2, orient=VERTICAL)
    global tree2
    tree2 = ttk.Treeview(frame2, columns=("Names"), height=200, selectmode="extended",
                    yscrollcommand=scrollbary.set)
    scrollbary.config(command=tree2.yview)
    scrollbary.pack(side=RIGHT, fill=Y)
    tree2.heading('Names', text="Names", anchor=W)
    #tree.heading('Fingerprints', text="Fingerprints", anchor=W)
    tree2.column('#0', stretch=NO, minwidth=0, width=0)
    #tree.column('#1', stretch=NO, minwidth=0, width=120)
    tree2.pack()

    mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database=values2[0]
    )

    mycursor = mydb.cursor()

    mycursor.execute("SHOW TABLES")
    db_list = []
    for x in mycursor:
        db_list.append(x[0])
    #print(x[0])

    for i in db_list:
        tree2.insert('', 'end', values=i)

    back_buttom= tk.Button(frame, text="Back", width=20, height=2, command=inside_db)#, command=Home
    back_buttom.place(relx=0.1,rely=0.815)
    
    delete_buttom= tk.Button(frame, text="Delete", width=20, height=2, command=call2)
    delete_buttom.place(relx=0.7,rely=0.815)

    add_modify_table= tk.Button(frame, text="Modify table", width=20, height=2, command=inside_db)
    add_modify_table.place(relx=0.5,rely=0.85, anchor="center")

def Home():
    for item in frame.place_slaves():
        item.place_forget()
    heading=tk.Label(frame,text='Welcome to the Database management tool',font=("Helvatical bold", 15),bg='lightblue')
    heading.place(relx=0.5,rely=0.1, anchor="center")

    #bt=tk.Button(root,text='page1',command=page1)
    #bt.grid(column=5,row=5)

    login_button= tk.Button(frame, text="Create Database", width=15, height=2, command=add_new_db)
    login_button.place(relx=0.6,rely=0.5)

    add_modify_button= tk.Button(frame, text="Existed Database", width=15, height=2, command=show_existing_db)
    add_modify_button.place(relx=0.3,rely=0.5)

    changeOnHover(login_button, "lightgreen", "white")
    changeOnHover(add_modify_button, "lightgreen", "white")


heading=tk.Label(frame,text='Welcome to the Database management tool',font=("Helvatical bold", 15),bg='lightblue')
heading.place(relx=0.5,rely=0.1, anchor="center")

#bt=tk.Button(root,text='page1',command=page1)
#bt.grid(column=5,row=5)

login_button= tk.Button(frame, text="Create Database", width=15, height=2, command=add_new_db)
login_button.place(relx=0.6,rely=0.5)

add_modify_button= tk.Button(frame, text="Existed Database", width=15, height=2, command=show_existing_db)
add_modify_button.place(relx=0.3,rely=0.5)

changeOnHover(login_button, "lightgreen", "white")
changeOnHover(add_modify_button, "lightgreen", "white")



root.mainloop()
```
